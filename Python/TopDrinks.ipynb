#### Tratamiento de Valores Nulos

```python
DataPurchases[DataPurchases['Description'].str.contains('Pinnacle Rainbow Sherbet')] #llenar valor nulo con dato 750mL
```

```python
# Filtrar las filas donde 'Description' es igual a 'Pinnacle Rainbow Sherbet' y 'Size' es NaN, y establecer 'Size' como '750mL'
DataPurchases.loc[(DataPurchases['Description'] == 'Pinnacle Rainbow Sherbet') & (DataPurchases['Size'].isna()), 'Size'] = '750mL' #se llenan los NaN 
```

```python
DataPurchases[DataPurchases['Description'].str.contains('Skinnygirl Pina Colada')] #llenar valor nulo con dato 750mL
```

```python
DataPurchases.loc[(DataPurchases['Description'] == 'Skinnygirl Pina Colada') & (DataPurchases['Size'].isna()), 'Size'] = '750mL' #se llenan los NaN 
```

```python
DataPurchases[DataPurchases['Description'].str.contains('Alabaster 07 Tinta de Toro')]
```

```python
DataInvoice.drop(columns=['Approval'], inplace=True) #eliminamos columna Approval porque mas del 90% de registros son nulos
```

```python
DataInventFinal[DataInventFinal['Store'].astype(str).str.contains('46')]
```

```python
DataInventInicial[DataInventInicial['Store'].astype(str).str.contains('46')] #En la tabla InventarioInicial verificamos que la tienda 46 corresponde a "TYWARDREATH"
```

```python
# Filtrar las filas donde 'Store' es igual a 46 y 'City' es NaN, y establecer 'City' como 'TYWARDREATH'
DataInventFinal.loc[(DataInventFinal['Store'] == 46) & (DataInventFinal['City'].isna()), 'City'] = 'TYWARDREATH' #se llenan los NaN 
```

```python
DataInventFinal[DataInventFinal['Store'].astype(str).str.contains('46')] #se verifica que la tabla ya no tenga valores nulos
```
